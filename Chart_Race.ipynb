```python
import pandas as pd
import bar_chart_race as bcr
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML

sales = pd.read_csv("/content/Sales MoM '21-24.csv")
sales.tail()
```

```python
sales = sales[sales['Month']!= 'Grand Total']
sales.tail()
```

```python
# drop column unmapped
sales.drop(['Unmapped'], axis='columns', inplace=True)
sales.head()
```

```python
#  datetime format of dayfirst
sales['Month'] = pd.to_datetime(sales['Month'], dayfirst=True)
```

```python
# Convert 'date' column to datetime
sales['Month'] = pd.to_datetime(sales['Month'])

# Set 'date' as the index
sales.set_index('Month', inplace=True)
```

```python
sales.dtypes
```

```python
bcr.bar_chart_race(
    df=sales,
    filename='sales_trend.mp4',
    orientation='h',
    sort='desc',
    n_bars=16,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=20,
    interpolate_period=True,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18, 's': f'Total sales: {v.nlargest(16).sum():,.0f}', 'ha': 'right', 'size': 4, 'family': 'DejaVu Sans'},
    # perpendicular_bar_func='median',
    period_length = 2500,
    figsize=(5, 3),
    dpi=240,
    cmap='dark12',
    title='MoM Sales per Region',
    title_size='',
    bar_label_size=5,
    tick_label_size=5,
    shared_fontdict={'family': 'DejaVu Sans', 'color': '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False
)
```
